```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sal=pd.read_csv('G:\Data Analysis\Salary Data_Sharath.csv')
pt=pd.read_csv('G:\Data Analysis\Salary Data_Sharath.csv')
```

```python
plt.figure(figsize=(12, 8))

sal_corr = sal.corr()
sns.heatmap(sal_corr, 
            xticklabels = sal_corr.columns.values,
            yticklabels = sal_corr.columns.values,
            annot = True);
```

```python
sns.boxplot(x='Gender',y='CTC',data=sal )
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x='GradYear',y='CTC',data=sal )
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.violinplot(x='GradYear',y='CTC',data=sal )
plt.xticks(rotation=90)
plt.show()
```
